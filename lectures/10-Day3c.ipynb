1. create a function that creates a gaussian bell curve.$g(x)=\frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}((x-\mu)/\sigma)^2}$ 

```python
#Function
def gauss(x=None,mu=0,sigma=1):
    scale=sigma*np.sqrt(2*np.pi)
    expo=(x-mu)**2/(-2*sigma**2)
    return np.e**expo/scale
```

```python
dicten={}
x=np.arange(start=-5,stop=6+0.1,step=0.1)
for sigma in np.arange(0.5,2+0.3,0.3):
    stringen='mu={},sigma={:2g}'.format(1,sigma)
    dicten[stringen]=gauss(x,mu=1,sigma=sigma)
import pandas as pd
df=pd.DataFrame(dicten,index=x)
df.plot()
```

Advanced:
use np.meshgrid to make a symmetric two dimensional coordinate grid and create a function that calculates the two dimensional gauss curve and plot it with plt.pcolormesh
now use this matrix as filter for the pictures from the first part of the lecture, what do you see?

```python
x=np.arange(-5,5,0.1)
X,Y=np.meshgrid(x,x)
def gauss2d(X,Y,mu=(0,0),sigma=(1,1)):
    scale=(sigma[0]+sigma[1])*(2*np.pi)
    expo=(X-mu[0])**2/(-2*sigma[0]**2)+(Y-mu[1])**2/(-2*sigma[1]**2)
    return np.e**expo/scale
print(X[:5,:5],'\n',Y[:5,:5])
```

```python
plotted_stuff=plt.pcolormesh(X,Y,gauss2d(X,Y))
fig=plt.gcf();ax=plt.gca() #this is where things get dirty, we learn more later
fig.colorbar(plotted_stuff,ax=ax)

from scipy import misc,ndimage
coins= misc.ascent()
x=range(0,20)
X,Y=np.meshgrid(x,x)
g=gauss2d(X,Y,sigma=(9,9))
coins1=ndimage.convolve(coins,g,mode='constant',cval=0)
sho(coins1)
```
